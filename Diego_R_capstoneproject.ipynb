```python
# Drop specified columns High correlated features
columns_to_drop = ['GenHlth', 'DiffWalk', 'Education']
df_processed = df.drop(columns=columns_to_drop)
```

```python
# Separate features and target variable
X = df_processed.drop('Diabetes_012', axis=1)
y = df_processed['Diabetes_012']
```

```python
# Apply SMOTE to the entire dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

##### Building the models

```python
# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Identify categorical and numerical columns
categorical_columns = X.select_dtypes(include=['object', 'category']).columns
numerical_columns = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Create a preprocessing pipeline
transformers = [('num', StandardScaler(), numerical_columns)]

if len(categorical_columns) > 0:
    transformers.append(('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_columns))

preprocessor = ColumnTransformer(transformers=transformers)
```

```python
# Create classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB()
}

results = {}

for name, classifier in classifiers.items():
    # Create a pipeline with preprocessor and classifier
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', classifier)
    ])
    
    # Fit the pipeline
    pipeline.fit(X_train, y_train)
    
    # Make predictions on train and test sets
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)
    
    # Calculate metrics for train and test sets
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred)
    train_recall = recall_score(y_train, y_train_pred)
    train_f1 = f1_score(y_train, y_train_pred)
    train_auc = roc_auc_score(y_train, y_train_pred)
    
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred)
    test_recall = recall_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    test_auc = roc_auc_score(y_test, y_test_pred)
    
    # Create confusion matrices
    train_cm = confusion_matrix(y_train, y_train_pred)
    test_cm = confusion_matrix(y_test, y_test_pred)
    
    results[name] = {
        'Train': {
            'Accuracy': train_accuracy,
            'Precision': train_precision,
            'Recall': train_recall,
            'F1 Score': train_f1,
            'AUC': train_auc,
            'Confusion Matrix': train_cm
        },
        'Test': {
            'Accuracy': test_accuracy,
            'Precision': test_precision,
            'Recall': test_recall,
            'F1 Score': test_f1,
            'AUC': test_auc,
            'Confusion Matrix': test_cm
        }
    }
```

```python
# Print results and display confusion matrices
for name, metrics in results.items():
    print(f"\n{name}:")
    for set_name in ['Train', 'Test']:
        print(f"  {set_name} Set:")
        for metric, value in metrics[set_name].items():
            if metric != 'Confusion Matrix':
                print(f"    {metric}: {value:.4f}")
        
        # Display confusion matrix
        cm = metrics[set_name]['Confusion Matrix']
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Diabetes', 'Diabetes'])
        disp.plot(cmap='Blues')
        plt.title(f'{name} - {set_name} Set Confusion Matrix')
        plt.show()
```

```python
# Visualize ROC curves for test set
plt.figure(figsize=(10, 8))
for name, classifier in classifiers.items():
    # Create and fit a new pipeline for each classifier
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', classifier)
    ])
    pipeline.fit(X_train, y_train)
    
    # Get predictions and calculate ROC curve
    y_pred_proba = pipeline.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    auc = roc_auc_score(y_test, y_pred_proba)
    
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.2f})')

plt.plot([0, 1], [0, 1], linestyle='--', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves (Test Set)')
plt.legend()
plt.show()
```

The Random Forest model is the top performer, followed by Gradient Boosting and KNN.

The ROC curves suggest that the tree-based models (Random Forest and Gradient Boosting) are particularly effective for this dataset.

Models like Logistic Regression and Naive Bayes, while still performing above random, do not perform as well as the tree-based methods in this scenario.

```python
# Create empty lists to store the results
model_names = []
combined_metrics = []

# Collect results
for name, metrics in results.items():
    model_names.append(name)
    
    # Combine train and test metrics
    combined_metric = {
        'train_accuracy': metrics['Train']['Accuracy'],
        'test_accuracy': metrics['Test']['Accuracy'],
        'train_precision': metrics['Train']['Precision'],
        'test_precision': metrics['Test']['Precision'],
        'train_recall': metrics['Train']['Recall'],
        'test_recall': metrics['Test']['Recall'],
        'train_f1': metrics['Train']['F1 Score'],
        'test_f1': metrics['Test']['F1 Score'],
        'train_auc': metrics['Train']['AUC'],
        'test_auc': metrics['Test']['AUC']
    }
    combined_metrics.append(combined_metric)

# Create DataFrame
combined_df = pd.DataFrame(combined_metrics, index=model_names)

# Rename the index
combined_df.index.name = 'Model'

# Reset index to make 'Model' a column
combined_df = combined_df.reset_index()

# Reorder columns if needed
column_order = ['Model', 
                'train_accuracy', 'test_accuracy',
                'train_precision', 'test_precision',
                'train_recall', 'test_recall',
                'train_f1', 'test_f1',
                'train_auc', 'test_auc']
combined_df = combined_df[column_order]

# Display the DataFrame
combined_df
```

**Improving the best model (Random Forest)**

- Hyperparameter tuning

```python
%%time
# Hyperparameter tuning with RandomizedSearchCV

# Define the parameter distribution for random search
param_dist = {
    'n_estimators': randint(40, 50),
    'max_depth': randint(15, 20),
    'min_samples_split': randint(2, 4),
    'min_samples_leaf': randint(1, 2),
    'max_features': uniform(0, 1)  # uniform distribution between 0 and 1
}

# Create a Random Forest classifier
rf = RandomForestClassifier(random_state=42)

# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=100,  # number of parameter settings that are sampled
    cv=5,  # 5-fold cross-validation
    verbose=2,
    random_state=42,
    n_jobs=-1  # use all available cores
)

# Fit RandomizedSearchCV
random_search.fit(X_train, y_train)

# Get the best parameters
best_params = random_search.best_params_
print("Best parameters found: ", best_params)
```

```python
# Evaluate the tuned model
best_rf = random_search.best_estimator_

# Make predictions on train and test sets
y_train_pred = best_rf.predict(X_train)
y_test_pred = best_rf.predict(X_test)

# Calculate metrics
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
train_precision = precision_score(y_train, y_train_pred, average='macro')
test_precision = precision_score(y_test, y_test_pred, average='macro')
train_recall = recall_score(y_train, y_train_pred, average='macro')
test_recall = recall_score(y_test, y_test_pred, average='macro')
train_f1 = f1_score(y_train, y_train_pred, average='macro')
test_f1 = f1_score(y_test, y_test_pred, average='macro')

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Training Precision: {train_precision:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Training Recall: {train_recall:.4f}")
print(f"Test Recall: {test_recall:.4f}")
print(f"Training F1 Score: {train_f1:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
```

```python
# Calculate probabilities for ROC curve
y_train_proba = best_rf.predict_proba(X_train)[:, 1]
y_test_proba = best_rf.predict_proba(X_test)[:, 1]

# Calculate metrics
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
train_precision = precision_score(y_train, y_train_pred, average='macro')
test_precision = precision_score(y_test, y_test_pred, average='macro')
train_recall = recall_score(y_train, y_train_pred, average='macro')
test_recall = recall_score(y_test, y_test_pred, average='macro')
train_f1 = f1_score(y_train, y_train_pred, average='macro')
test_f1 = f1_score(y_test, y_test_pred, average='macro')
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_proba)
roc_auc_train = auc(fpr_train, tpr_train)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_proba)
roc_auc_test = auc(fpr_test, tpr_test)

# Print metrics
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Training Precision: {train_precision:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Training Recall: {train_recall:.4f}")
print(f"Test Recall: {test_recall:.4f}")
print(f"Training F1 Score: {train_f1:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
print(f"Training ROC-AUC: {roc_auc_train:.4f}")
print(f"Test ROC-AUC: {roc_auc_test:.4f}")
```

**Approach 2** Identifying collinearity through Variance Inflation Factor (VIF) and applying Principal Component Analysis (PCA) to create new vectors to use them to train and test the ML algorithms.

```python
# Checking Collinearity
# Drop the target variable for VIF calculation
X = df.drop(columns='Diabetes_012')

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

# Sort the dataframe by VIF values in descending order
vif_data = vif_data.sort_values('VIF', ascending=False).reset_index(drop=True)

vif_data
```

##### Interpreting the Results

**VIF < 5**: No multicollinearity problem.

**5 <= VIF < 10**: Moderate multicollinearity.

**VIF >= 10**: High multicollinearity, problematic.

**features with high VIF values**:

Education: 29.507416

CholCheck: 23.187436

AnyHealthcare: 20.839710

BMI: 18.149913

Income: 14.156118

GenHlth: 10.740162

Having this into account, we will perfom Principal Components Analysis. It creates new uncorrelated variables from the original ones so collinearity is eliminated, which could be a problem for some machine learning algorithms.

```python
# Separate binary, categorical, and numerical features
binary_features = ['HighBP', 'HighChol', 'Smoker', 'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk', 'Sex']
categorical_features = ['Education', 'Income']
numerical_features = ['BMI', 'GenHlth', 'MentHlth', 'PhysHlth', 'Age']

# One-hot encode categorical features
categorical_transformer = OneHotEncoder(drop='first', sparse_output=False)
encoded_categorical = categorical_transformer.fit_transform(df[categorical_features])

# Combine all features
X = np.hstack((df[binary_features].values, encoded_categorical, df[numerical_features].values))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Initialize PCA
pca = PCA(n_components=0.95)  # Retain 95% of the variance
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)

# Plot cumulative explained variance
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(np.arange(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Principal Components')
plt.grid()
plt.show()
```

##### Analysis of PCA Results

- Explained Variance Ratio: Each principal component explains a portion of the variance in the data. The first principal component explains the highest amount of variance, followed by the second, and so on. The explained variance ratio decreases as the number of principal components increases.

- Cumulative Explained Variance: The cumulative explained variance plot shows how the total variance explained by the principal components increases as more components are added. This plot helps in determining the number of principal components that should be considered to capture most of the variance in the data.

Significant Principal Components: From the cumulative explained variance plot, we observe that the first few principal components explain a substantial amount of variance. For instance, the first 5 principal components explain about 60% of the variance.

Dimensionality Reduction: To reduce the dimensionality while retaining most of the variance, we can select the number of principal components that cumulatively explain a desired threshold (e.g., 90%). In this case, approximately 15 principal components are needed to explain 90% of the variance.

```python
%%time
# Extract models and parameters from results
models_and_params = results_df_sorted['params']

# Initialize an empty list to store results
results = []

# Process each model's parameters and evaluate the model
for i, params in enumerate(models_and_params):
    model = clone(pipeline)
    
    # Convert string representation of params to dictionary
    param_dict = eval(params)
    
    # Set the classifier and its parameters in the pipeline
    model.set_params(**param_dict)
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    
    # Predictions on the training set
    y_train_pred = model.predict(X_train)
    y_train_pred_proba = model.predict_proba(X_train)[:, 1]
    
    # Predictions on the test set
    y_test_pred = model.predict(X_test)
    y_test_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # Calculate evaluation metrics for the training set
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred, average='weighted')
    train_recall = recall_score(y_train, y_train_pred, average='weighted')
    train_f1 = f1_score(y_train, y_train_pred, average='weighted')
    train_roc_auc = roc_auc_score(y_train, y_train_pred_proba, multi_class='ovr')
    
    # Calculate evaluation metrics for the test set
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred, average='weighted')
    test_recall = recall_score(y_test, y_test_pred, average='weighted')
    test_f1 = f1_score(y_test, y_test_pred, average='weighted')
    test_roc_auc = roc_auc_score(y_test, y_test_pred_proba, multi_class='ovr')
    
    # Append the results
    results.append({
        'model': str(params),
        'train_accuracy': train_accuracy,
        'test_accuracy': test_accuracy,
        'train_precision': train_precision,
        'test_precision': test_precision,
        'train_recall': train_recall,
        'test_recall': test_recall,
        'train_f1': train_f1,
        'test_f1': test_f1,
        'train_roc_auc': train_roc_auc,
        'test_roc_auc': test_roc_auc
    })
    
    # Confusion matrix for the test set
    conf_matrix = confusion_matrix(y_test, y_test_pred)
    
    # Plot confusion matrix heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'Confusion Matrix for Model {i+1}')
    plt.show()

# Convert results to DataFrame for better visualization
results_df = pd.DataFrame(results)

# Display the results
results_df
```

RandomForestClassifier: Despite being the top performer in terms of training metrics, it shows significant overfitting, indicating room for improvement in terms of generalization. The overfitting here is bigger that the one we got from the approach 1.

KNeighborsClassifier: Offers a balance between performance and generalization, making it a strong contender for this dataset.

GradientBoostingClassifier: Demonstrates good stability and moderate performance, making it a reliable choice with less risk of overfitting.

LogisticRegression: The most consistent model, with almost identical performance on both train and test sets, though it’s not the most powerful classifier.

GaussianNB: The weakest performer, showing low accuracy and ROC-AUC scores, but consistent in its results.

The key takeaway is that Random Forest is the most powerful but also the most overfit, while models like KNN and Logistic Regression are more stable and generalize better, though with slightly lower performance. So Our next step is improve Random Forest through hyperparameter tuning.
